```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cancer_data = pd.read_csv("Cancer_data (1).csv")
cancer_data
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = cancer_data['perimeter_mean'].quantile(0.25)
Q3 = cancer_data['perimeter_mean'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = cancer_data[(cancer_data['perimeter_mean'] < lower_bound) | (cancer_data['perimeter_mean'] > upper_bound)]
print(f'Number of outliers: {len(outliers)}')
print(outliers['perimeter_mean'])
```

```python
# Plot the boxplot to visualize outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=cancer_data['perimeter_mean'])
plt.title('Boxplot of Perimeter Mean')
plt.xlabel('Perimeter Mean')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import numpy as np
```

```python
cancer_data = pd.read_csv("Cancer_data (1).csv")
cancer_data
```

```python
# Extract the perimeter_mean column
perimeter_mean = cancer_data['perimeter_mean']

# Calculate the KDE
kde = gaussian_kde(perimeter_mean)

# Define a range of values for perimeter_mean
x_values = np.linspace(min(perimeter_mean), max(perimeter_mean), 1000)

```

```python
# Calculate the PDF values
pdf_values = kde(x_values)

# Calculate the CDF values
cdf_values = np.cumsum(pdf_values) * (x_values[1] - x_values[0])

# Find the probability for perimeter_mean <= 20
probability_index = np.where(x_values <= 20)[0]
if len(probability_index) > 0:
    probability = cdf_values[probability_index[-1]]
else:
    probability = 0
print(f'Probability of perimeter_mean <= 20: {probability}')
```

```python
# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(x_values, cdf_values, label='CDF of Perimeter Mean')
plt.axvline(x=20, color='r', linestyle='--', label='Perimeter Mean = 20')
plt.xlabel('Perimeter Mean')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of Perimeter Mean')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
cancer_data = pd.read_csv("Cancer_data (1).csv")
cancer_data
```

```python
perimeter_mean = cancer_data['perimeter_mean']
area_mean = cancer_data['area_mean']
```

```python
# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=perimeter_mean, y=area_mean)
plt.title('Scatter Plot of Perimeter Mean vs Area Mean')
plt.xlabel('Perimeter Mean')
plt.ylabel('Area Mean')
plt.show()
# Calculate the Pearson correlation coefficient
correlation, p_value = pearsonr(perimeter_mean, area_mean)
print(f'Pearson correlation coefficient: {correlation}')
print(f'P-value: {p_value}')
```
