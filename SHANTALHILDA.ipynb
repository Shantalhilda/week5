### Question 1
#### Probability of a student being a student being a graduate under the variable "Target" ?


```python
#probability of a student being a graduate under the variable "Target"
grad_prob = shantal['Target'].value_counts(normalize=True).get('Graduate', 0)
grad_prob
```

### Question 2
#### Probability of a student being a "Dropout" with a Martial Status of 1 ?


```python
#probability of a student being a "Dropout" with a Martial Status of 1
dropout_prob =shantal[(shantal["Target"] == "Dropout") & (shantal["Marital status"] == 1)]
dropout_prob = len(dropout_prob)/len(shantal)
dropout_prob
```

```python
#Relationship between "Target" and "Mother's Qualification"
shantal.groupby(["Target", "Mother's qualification"]).size().unstack().plot(kind="bar", stacked=True)
plt.show()
```
